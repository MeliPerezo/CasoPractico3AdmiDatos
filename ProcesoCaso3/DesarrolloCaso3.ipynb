8. Eliminar aquella columna (s) que tengan más del 70% de sus valores en nulo.

```python
for x in DataFrame: 
    y= DataFrame[x].isna().mean()
    if y>0.7 :
        DataFrame.drop(x, inplace=True, axis=1)
```

```python
print(len(DataFrame))
```

9. Eliminar aquellas filas que tenga menos de dos elementos.

```python
cols=DataFrame.columns

for i in range(0, len(DataFrame)):
    y=DataFrame.loc[i,cols].isna().sum()
    if y>2:
        DataFrame=DataFrame.drop([i])
```

```python
DataFrame
```

10. Reemplace los String “--" por valores nulos

```python
DataFrame.replace(to_replace="--", value=np.nan, inplace=True)
```

```python
DataFrame
```

11. Reemplazar los valores nulos de la columna de “Gender” por U (Unknown) para
aquellos empleados que no especificaron este campo.

```python
DataFrame["Gender"].replace(to_replace=np.nan, value="Unknown", inplace=True)
```

```python
DataFrame[DataFrame["Gender"]=="Unknown"]
```

12. Reemplazar los valores nulos de la columna “Departamento” por “Bench”.

```python
DataFrame["Department"].replace(to_replace=np.nan, value="Brench", inplace=True)
```

```python
DataFrame[DataFrame["Department"]=="Brench"]
```

13. Reemplazar los valores nulos de la columna “HourlyRate” por la mediana.

```python
DataFrame["HourlyRate"].replace(to_replace=np.nan, value=DataFrame["HourlyRate"].median(), inplace=True)
```

```python
DataFrame[DataFrame["HourlyRate"]==DataFrame["HourlyRate"].median()]
```

14. Reemplazar los valores perdidos inesperados de la columna “Overtime” por No.

```python
DataFrame["OverTime"].replace(to_replace=[np.nan, "0"], value="No", inplace=True)
```

```python
DataFrame["OverTime"].replace(to_replace=[np.nan, "1"], value="Yes", inplace=True)
```

```python
DataFrame["OverTime"].unique()
```
